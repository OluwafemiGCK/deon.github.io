```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic function
def logistic(x):
    return 1 / (1 + np.exp(-x))

# Coefficients
gamma = np.array([3.0651, 0.5293, 0.4240, 0.5334, -2.2511, 0.2537])

# Student's characteristics
GCSE = 6
Male = 1
Private = 0
Maths = 1

# Compute logit
logit_p = gamma[0] + gamma[1]*Male + gamma[2]*Private + gamma[3]*Maths + gamma[4]*GCSE + gamma[5]*GCSE**2
logit_p = gamma[0] + gamma[1]*Male + gamma[2]*Private + gamma[3]*Maths + gamma[4]*GCSE + gamma[5]*GCSE**2

# Compute probability using logistic function
p = logistic(logit_p)

# Create range of GCSE scores
GCSE_range = np.linspace(0, 10, 100)
logit_range = gamma[0] + gamma[1]*Male + gamma[2]*Private + gamma[3]*Maths + gamma[4]*GCSE_range + gamma[5]*GCSE_range**2
p_range = logistic(logit_range)

# Plot CDF
plt.figure(figsize=(10, 5))
plt.plot(GCSE_range, p_range, label='CDF', color='blue')
plt.axvline(x=GCSE, color='red', linestyle='--', label='Average GCSE Score')
plt.xlabel('Average GCSE Score')
plt.ylabel('Probability')
plt.title('Cumulative Distribution Function (CDF)')
plt.legend()
plt.grid(True)
plt.show()

# Plot PDF
plt.figure(figsize=(10, 5))
plt.plot(GCSE_range, np.gradient(p_range, GCSE_range[1]-GCSE_range[0]), label='PDF', color='green')
plt.axvline(x=GCSE, color='red', linestyle='--', label='Average GCSE Score')
plt.xlabel('Average GCSE Score')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF)')
plt.legend()
plt.grid(True)
plt.show()
```
